```python
import numpy as np                                                     #importing numpy
def bonus_func(sales_col):                                             # defining a function "bonus_func" with arg "sales_col"
    elig_values = np.where(sales_col > 10, "eligible","not eligible")  # "np.where", through this we are checking eligibility criteria for giving bonus to the stores where in jan month sale is greater than 10, will be consider as "eligible" else "not eligible".
    return elig_values
df["Jan_bonus"] = bonus_func(df["Jan"])                                # adding a new column "Jan_bonus" where we can see the stores those are eligible and those are not eligible acc. to Jan sales data.
df.head(10)
```

```python
# in the previous example 10 was fixed as an eligibility criteria. To define a variable criteria, like "Jan" sales compared with 15 in previous month and something else in next month, to do so..

def bonus_f(sales_column, sales_num):
    eligibility = np.where(sales_column > sales_num, "eligible", "not eligible")
    return eligibility
df["previous_crit"] = bonus_f(df["Jan"], 15)
df.head()
```

#### then we can also compare it by some other values because probably next month we want to compare the sales with a higher or smaller value than the previous month.


```python
df["next_crit"] = bonus_f(df["Jan"], 16)
df.head(10)
```

```python
# now, if we want to apply the same criteria to whole dataset..

def bonus(sales_col1):
    eligibility1 = np.where(sales_col1 > 10, "eligible", "not eligible")
    return eligibility1
newdf = df.select_dtypes(include = ['int64','float64']).apply(bonus)
newdf
```

```python
# creating a dataframe

years = [90, 91, 92, 93]
f1 = {90:8 , 91:9, 92:7, 93:10}
firm1 = pd.Series(f1, index = years)
firm1

f2= {90:13, 91:34, 94:11}
firm2 = pd.Series(f2, index = years)
firm2

df1 = pd.DataFrame(columns = ['Firm1','Firm2'], index = years)
df1

df1.Firm1 = firm1
df1.Firm2 = firm2
df1
```

```python
# if we want to drop NaN from df1
df1.dropna()   # dropping row wise

# for making this changes in original data..
#### df1.dropna(inplace = True)
```

```python
df1.dropna(axis = 1) #column wise
```

```python
threshdf = df1.dropna(axis =1, thresh = 3)  # keep only those columns which have at least 3 non-NAN values.

threshdf
```

```python
threshdf2 = df1.dropna(thresh = 2)  # keep only those rows which have at least 2 non-NAN values.
                                      # axis = 0(default), is for rows.
threshdf2
```

```python
# to fill all NaN values with 0

fillna1 = df1.fillna(0)
fillna1
```

```python
# to fill all NaN values with average of non-NaN values of that particular column(general method)..
fillna2 = df1.fillna(df1.mean())
fillna2
```
